```python
def clean_and_standardize_data(df, year, mapping, target_columns):
    df_modified = df.rename(columns = {k:v for k,v in mapping.items() if k in df.columns})
    df_modified['survey_year'] = year
    available_columns = [col for col in target_columns if col in df_modified.columns]
    return df_modified[available_columns + ['survey_year']]
    

df_2014_cleaned = clean_and_standardize_data(df_2014, 2014, column_mapping, target_columns)
df_2016_cleaned = clean_and_standardize_data(df_2016, 2016, column_mapping, target_columns)
df_2017_cleaned = clean_and_standardize_data(df_2017, 2017, column_mapping, target_columns)
df_2019_cleaned = clean_and_standardize_data(df_2019, 2019, column_mapping, target_columns)
df_2023_cleaned = clean_and_standardize_data(df_2023, 2023, column_mapping, target_columns)
```

#### Q. Does age influence whether someone seeks mental health treatment?

```python
# Plot the age distribution of different treatment group (Yes/No)
plt.figure(figsize=(12, 5))
sns.displot(data=df_age_filtered, x='age', col='treatment', kde=True)
plt.tight_layout()
```

```python
# binning the column "age" into age ranges
df_age_filtered['age_range'] = pd.cut(
    df_age_filtered['age'],
    [0, 25, 35, 45, 55, 65, 100],
    labels=['0-25', '26-35', '36-45', '46-55', '56-65', '66-100'],
    include_lowest=True
)

df_age_filtered.head()
```

```python
# plot the distribution of data within different age group for different treatment group (True/False)
treatment_per_age_range = df_age_filtered.groupby(
    ['age_range', 'treatment']
).size().unstack()

treatment_per_age_range
```

```python
# plot the distribution of data within different age group for different treatment group (True/False)
plt.figure(figsize=(10, 6))
treatment_per_age_range.plot(kind='bar')
plt.title('Treatment Distribution by Age Range')
plt.xlabel('Age Range', c='b')
plt.ylabel('Count', c='g')
plt.grid(axis='y')
plt.tight_layout()
plt.legend(title='Treatment')
```

Looking at the chart, there are clear diffrences in the distribution of data within different age group for different treatment group (True/False):
- 26-35 age group: Shows the largest disparity - much higher count for "Yes" treatment (~1,100) compared to "No" treatment (~850)
- 36-45 age group: Also shows higher "Yes" treatment (~590) than "No" treatment (~430)
- 0-25 age group: Shows roughly equal distribution between both treatment groups (~250 each)
- 46-55 age group: Shows slightly higher "Yes" treatment (~160) than "No" treatment (~100)
- 56-65 and 66+ age groups: Have very small sample sizes but still show some differences

The most pronounced difference is in the 26-35 and 36-45 age groups, where the "Yes" treatment group consistently has higher counts than the "No" treatment group. 


```python
from scipy.stats import chi2_contingency

# Run chi-square test for independence
chi2, p, dof, expected = chi2_contingency(treatment_per_age_range)

print("Chi-square statistic:", chi2)
print("Degrees of freedom:", dof)
print("p-value:", p)
print("\nExpected frequencies:\n", expected)
```

#### Interpretation
(H₀): Age and treatment status are independent (age has no influence).

(H₁): Age and treatment status are associated (age influences likelihood of treatment).

p < 0.05 → reject H₀ → age significantly influences treatment seeking

#### How strong is the association?

```python
import numpy as np

# Effect size 
n = treatment_per_age_range.values.sum() # total sample size
k = min(treatment_per_age_range.shape)   

cramers_v = np.sqrt(chi2 / (n * (k - 1)))
print("Cramér's V:", cramers_v)

```

#### Interpretation
The chi-square test (Chi²(5) = 21.68, p < 0.001) shows a statistically significant association between age group and treatment status. However, the effect size measured by Cramér's V (V = 0.076) indicates that the strength of this association is very small.

### Conclusion about Age and Treatment

- The distribution of treatment status varies across age groups, with the most pronounced differences in the **26–35** and **36–45** ranges, where treatment counts are higher.  
- However, this disparity is **largely due to the greater number of individuals** in these age groups.  
- The chi-square test showed a **significant association** between age group and treatment (**Chi²(5) = 21.68, p < 0.001**).  
- Cramér’s V (**0.076**) indicates that the **strength of the association is very weak**.  
- Therefore, while age and treatment are statistically related, **age is not a strong determinant of treatment**.  
- Other factors (beyond age) likely play a much larger role in determining treatment.

```python
def classify_gender(gender):
    if pd.isna(gender):
        return 'Other'

    val = str(gender).strip().lower() # normalize the gender
    
    if val in [s.lower().strip() for s in male_str]:
        return 'Male'
    elif val in [s.lower().strip() for s in female_str]:
        return 'Female'
    elif val in [s.lower().strip() for s in trans_str]:
        return 'Transgender'
    else:
        return 'Other'

df_gender_filtered['gender'] = df_gender_filtered['gender'].apply(classify_gender)
df_gender_filtered['gender'].value_counts(dropna=False).sort_values(ascending=False)
```

```python
# clean and map the care_options column
def map_care_options(care_options):
    if pd.isna(care_options):
        return "Uncertain"
    
    val = str(care_options).strip().lower()

    if val == "yes":
        return "Aware"
    elif val == "no":
        return "Not aware"
    else: # "not sure", "i am not sure", etc.
        return "Uncertain"

df_gender_filtered['care_options'] = df_gender_filtered['care_options'].apply(map_care_options)
df_gender_filtered['care_options'].value_counts(dropna=False)
```
